```python
# EDA 
```

```python
total_revenue = df["Sales"].sum()
print("Total Revenue : ",total_revenue)
```

```python
Category_wise_rev = (
    df.groupby("Category")["Sales"]
      .sum()
      .sort_values(ascending=False)
)

print(Category_wise_rev)
```

```python
top_5_cities = (
    df.groupby("City")["Sales"]
      .sum()
      .sort_values(ascending=False)
      .head(5)
)

print(top_5_cities)
```

```python
product_sales = (
    df.groupby("Product Name")["Sales"]
    .sum()
    .sort_values(ascending = False)
    .head(10)
)

print(product_sales)
```

```python
Category_Avg_profit = (
    df.groupby("Category")["Profit"]
      .mean()
      .sort_values(ascending=False)
      .head(1)
)

print(Category_Avg_profit)
```

```python
Highest_Category_Revenue = (
    df.groupby("Category")["Sales"]
      .sum()
      .sort_values(ascending=False)
      .head(1)
)

print(Highest_Category_Revenue)
```

```python
# Total company revenue
total_revenue = df["Sales"].sum()

# Category-wise revenue
category_revenue = df.groupby("Category")["Sales"].sum()

# Percentage contribution
category_revenue_percentage = (
    (category_revenue / total_revenue) * 100
).sort_values(ascending=False)

print(category_revenue_percentage.round(2))
```

```python
# City-wise revenue
city_revenue = df.groupby("City")["Sales"].sum()

# Percentage contribution
city_revenue_percentage = (
    (city_revenue / total_revenue) * 100
).sort_values(ascending=False).head(1)

print(city_revenue_percentage.round(2))
```

```python
Region_revenue = (
    df.groupby(["Region","Category"])["Sales"]
    .sum()
    .sort_values(ascending = False)
    .reset_index()
)

print(Region_revenue)
```

```python
category_summary = df.groupby("Category").agg(
    {
        "Sales":"sum",
        "Profit":"sum",
        "Discount":"mean",
        "Quantity":"sum"
    }
).sort_values(by="Sales", ascending=False)

print(category_summary)
```

```python
category_sales = (
    df.groupby("Category")["Sales"]
    .sum()
    .sort_values(ascending = False)
)
# print(category_sales)
```

```python
plt.figure(figsize=(12,5))
plt.xticks(rotation=45)
plt.bar(category_sales.index,category_sales.values)
plt.title("Category Sales")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

```python
plt.scatter(
    df["Discount"],
    df["Sales"]
)

plt.xlabel("Discount")
plt.ylabel("Sales")
plt.title("Discount vs Sales")
plt.show()
```

```python
month_order = [
    "January", "February", "March", "April",
    "May", "June", "July", "August",
    "September", "October", "November", "December"
]

df["Month"] = pd.Categorical(
    df["Month"],
    categories=month_order,
    ordered=True
)
```

```python
monthly_sales = (
    df.groupby("Month")["Sales"]
      .sum()
      .sort_index()      
)
```

```python
df.head()
```

```python
plt.figure(figsize=(12,5))

plt.bar(monthly_sales.index,monthly_sales.values)
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
plt.figure(figsize=(12,5))

plt.plot(monthly_sales.index,monthly_sales.values)
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
monthly = (
    df.groupby("Month")[["Sales","Profit"]]
    .sum()
)
print(monthly)
```

```python
plt.figure(figsize=(12,5))

plt.plot(monthly.index,monthly["Sales"],marker ="o")
plt.plot(monthly.index,monthly["Profit"],marker ="o")
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
category_summary = df.groupby("Category").agg({
    "Sales": "sum",
    "Profit": "sum"
})

category_summary["Profit Margin"] = (
    category_summary["Profit"] /
    category_summary["Sales"]
) * 100

print(category_summary)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr_matrix = df.corr(numeric_only=True)

plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```
